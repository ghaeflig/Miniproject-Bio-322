## Correlation

```r
pairs(data[1:10])
```

By running a pairs plot on some predictors we can see that some predictors are well correlated :
- For example, Sv, SE and Sp
- It seems to be good to search for correlation equal to 1 between to predictors

```r
datac <- data[,-which(names(data) %in% c("VALENCE.PLEASANTNESS"))]
dim(datac)
cor_matrix <- cor(datac)
n = 0
cor_i <- c()
cor_j <- c()
for(i in 1:nrow(cor_matrix)) {
    for(j in 1:i) #because the correlation matrix is symetric
        if(i != j) {
            if(cor_matrix[i,j] == 1) {
                n = n + 1
                cor_i <- c(cor_i,i)
                cor_j <- c(cor_j,j)
            }   
        }          
}
n
cor_i <- c(cor_i,0)
filtered_cor_i <- c()
for(i in 1:n) {
    if(cor_i[i] != cor_i[i+1]) filtered_cor_i <- c(filtered_cor_i,cor_i[i])
}
length(filtered_cor_i)
filtered_data <- datac
filtered_data <- filtered_data[,-filtered_cor_i]
dim(filtered_data)

#--> We found 352 correlations that are equal to 1 and when we count the predictors that are many times correlated only
# one time, we can put off 147 predictors 
```

By checking pairs correlations between predictors thanks to the correlation matrix :
- We found 352 correlations that are equal to 1
- When we count the predictors that are many times correlated only one time, we can remove 147 predictors 

However, by using the filtered data in model predictions we saw :
- For lasso and ridge prediction, this data increase a little the test error (of the order of the 3 or 4 digits after the decimal point)
- So, we decided not to take the filtered data in model prediction
